### Transposing a matrix

````python
# In native python

# Create a 2D "array" using a list of lists
nv = [
    [1, 2, 3, 4, 5],
    [6, 7, 8, 9, 0]
]

# For a matrix with R rows and C columns, the transpose has C rows and R columns.

def transpose(mat):

  print(f"vector: {mat}\n")

  rows = len(mat)            # number of rows
  cols = len(mat[0])         # number of columns (assumes all rows same length)
  print(f"rows: {rows}")
  print(f"columns: {cols}\n")

  nvT = []                  # this will store the transposed matrix

  for c in range(cols):
      new_row = []
      for r in range(rows):
          new_row.append(mat[r][c])
          # print(f"mat[{r}][{c}]: {mat[r][c]}")
      # print(f"new row: {new_row}")
      nvT.append(new_row)

  print(f"transpose: {nvT}\n")
  rows = len(nvT)            # number of rows
  cols = len(nvT[0])         # number of columns (assumes all rows same length)
  print(f"new rows: {rows}")
  print(f"new columns: {cols}\n")

  return nvT

nvT = transpose(nv)

````

### Softmax

The softmax function converts raw prediction scores (logits) from a neural network into a probability distribution. For a vector of logits $z=[z_1,z_2,...,z_K]$, softmax is mathematically defined as: [geeksforgeeks](https://www.geeksforgeeks.org/deep-learning/the-role-of-softmax-in-neural-networks-detailed-explanation-and-applications/)

$Softmax(z_i)=\frac{e^{z_i}}{∑_{j=1}^Ke^{z_j}}$

The exponential in the numerator ensures all outputs are positive, while the denominator normalizes them so probabilities sum to 1. This works with any real-valued logits—positive, negative, or zero. [deepai](https://deepai.org/machine-learning-glossary-and-terms/softmax-layer)

````python
# import libraries

import numpy as np
import torch as t
import torch.nn as nn
import matplotlib.pyplot as plt
import warnings
````

````python
# manually in numpy

z = [1,2,3]

num = np.exp(z)
den = np.sum(np.exp(z))
softmax = num/den

print(f"Softmax of z: {softmax}")
print(f"Sum of Softmax of z: {np.sum(softmax)}")
````

````python
# manually in numpy
warnings.filterwarnings('ignore')

z = np.random.randint(-25,25,size=10)
print(f"z: {z}")

num = np.exp(z)
den = np.sum(np.exp(z))
softmax = num/den

print(f"Softmax of z: {np.round(softmax,3)}")
print(f"Sum of Softmax of z: {np.round(np.sum(softmax),0)}")

# compare
plt.plot(z,softmax,'ko')
plt.xlabel("Original Number (z)")
plt.ylabel("Softmax $\sigma$")
# plt.yscale('log')
plt.show()
````

### Log Function

Why log function is preferred in a loss function?
    
1. **Numerical Stability:** Log functions prevent **overflow and underflow issues** common in deep learning. Without logs, computing products of small probabilities (e.g., 0.001×0.002×0.00010.001×0.002×0.0001) underflows to zero. The logarithm converts products to sums:  log⁡(p1×p2×p3)=log⁡(p1)+log⁡(p2)+log⁡(p3), maintaining numerical precision. Computing loss from logits (raw model outputs) before applying softmax further improves stability by avoiding intermediate probability calculations that can overflow

2. **Lower Values give Higher Differences:** The closer the log function gets to input zero, it gives bigger jumps to small movements which helps during training when improvements comes down in last stages where improvements are marginal or loss are small

### Entropy and Cross-Entropy



````python
# import libraries
import numpy as np
import torch as t
import matplotlib.pyplot as plt
````

````python
# entropy
p = 0.25
q = 1 - p

x = [p,q]
H = 0
for i in x:
  H -= i*np.log(i)

print(f"Correct Entropy take both p and (1-p = q) : {H.round(3)}")
````

````python
# cross entropy
p = [0, 1]
q = [0.25, 0.75]

H = 0
for i in range(len(p)):
  H -= p[i]*np.log(q[i])

print(f"Cross Entropy between Ground Truth p and Prediction q: {round(H,3)}")
````

````python
import torch.nn.functional as F

tensorP = t.Tensor(p).float()
tensorQ = t.Tensor(q).float()

x = F.binary_cross_entropy(tensorQ,tensorP).item()
print(f"Cross Entropy in PyTorch: {round(x,3)}")
````

### Argmin and Argmax

Argmin (argument of the minimum) is a function that returns the index of the smallest element in an array or tensor, rather than the value itself.

**The key distinction:** argmax/argmin are non-differentiable operations (they produce discrete indices), making them unsuitable for direct backpropagation. This is why straight-through estimators or relaxations (like softmax temperatures) are used when gradient flow through discrete choices is needed.

````python
# import libraries
import numpy as np
import torch as t
````

````python
# repeat with matrix
M = np.array([ [0,1,10], [20,8,5] ])
print(M), print(' ')

# various minima in this matrix!
minvals1 = np.min(M)        # minimum from ENTIRE matrix
minvals2 = np.min(M,axis=0) # minimum of each column (across rows)
minvals3 = np.min(M,axis=1) # minimum of each row (across columns)

# print them out
print(minvals1)
print(minvals2)
print(minvals3)
````

````python
# various minima in this matrix!
minidx1 = np.argmin(M)        # minimum from ENTIRE matrix
minidx2 = np.argmin(M,axis=0) # minimum of each column (across rows)
minidx3 = np.argmin(M,axis=1) # minimum of each row (across columns)

# print them out
print(M), print(' ') # reminder
print(minidx1)
print(minidx2)
print(minidx3)
````

### Mean and Variance

Mean is the arithmetic average—the sum of all values divided by their count. It represents the central location of a dataset.​

Variance is the average squared deviation from the mean. Mathematically:

$σ^2 = \frac{∑(Xi−μ)^2}{N-1}$, where $X_i$  are individual values,
μ is the mean, and N is the count.​

------
**Where Mean Fails**

The mean's critical limitation is its extreme sensitivity to outliers. In skewed distributions or datasets with extreme values, the mean gets pulled toward the outliers and no longer represents the "typical" value well. For example, in a salary dataset where one executive earns 10 million and five employees earn 50,000 each, the mean (~$1.85M) misrepresents the typical salary.
Additionally, the mean alone provides no information about data spread—two datasets with identical means can have wildly different distributions.​

------
**Why Squared Deviation vs. Absolute Deviation**

Using squared differences instead of absolute values offers several mathematical and practical advantages:

1. Differentiability & Mathematical tractability: Squared deviations are continuously differentiable everywhere, while absolute values have a non-differentiable point at zero. This matters for optimization, calculus operations, and computational algorithms.​

2. Additivity for independent variables: Variances of independent random variables are additive — $Var(X+Y)=Var(X)+Var(Y)$ — a property that mean absolute deviations lack. This is invaluable in probability theory and statistical inference.​

3. Penalizes outliers more heavily: Squaring amplifies large deviations (e.g., a deviation of 50 becomes 2,500, not 50), making variance more sensitive to extreme values. This is useful when outliers signal important phenomena.​​

**In essence:** squared deviations work with the underlying mathematics of statistics, while absolute deviations, though more intuitive, don't compose well in complex calculations.​

````python
import numpy as np

N = 100000
x = np.random.randint(0,high=100,size=N)

mean = np.mean(x)
var0 = np.var(x,ddof=0) # default
var1 = np.var(x,ddof=1) # unbiased

print(mean)
print(var0)
print(var1)
print("For large set of data the degree of freedom has very low impact")
````

### Random Sampling and Reproducible Sampling with Seeding

Random sampling involves selecting subsets of data from a larger dataset, with the selection process governed by pseudo-random number generators (PRNGs). In deep learning, randomness is intrinsic to many operations—weight initialization, data shuffling, dropout, batch composition, and hyperparameter selection all rely on random number generation.

**Reproducible sampling with seeding** addresses a critical challenge:

Controlling this randomness to ensure consistent, verifiable results. By setting a fixed seed value, you initialize the PRNG to generate an identical sequence of "random" numbers across multiple executions. When a seed is set (e.g., random.seed(42)), subsequent random operations produce the same output every time—enabling exact reproducibility.

**Where Randomness Occurs in Deep Learning**

Weight initialization uses randomness to break symmetry in neural networks. Data shuffling and batch composition depend on random sampling. Dropout stochastically removes neurons during training. Optimization algorithms may use randomized hyperparameter searches. Even train-test splitting requires controlled randomization.​​

**Importance for Deep Learning**

Reproducibility is essential for scientific validity and debugging. Uncontrolled randomness makes it impossible to isolate whether performance differences stem from architecture changes or stochastic variation. Research shows that only 3% of deep learning bugs are reproducible without deliberate seed management, and different random seeds produce significantly different model accuracy—sometimes with 11-31% variation on benchmark datasets.

When random seeds are not fixed, model performance becomes unreliable; changing hardware, library versions, or execution order produces different results. This undermines trust in production systems and clinical applications where consistency is paramount.

**Practical Implementation**

Set seeds across all PRNG sources simultaneously:

```
import random
import numpy as np
import torch

random.seed(42)
np.random.seed(42)
torch.manual_seed(42)
if torch.cuda.is_available():
  torch.cuda.manual_seed_all(42)
```


For DataLoaders in PyTorch, use `worker_init_fn` and generator parameters to maintain reproducibility across multiprocessing workers.

**Caveat**:

Identical seeds produce identical random sequences, but different seeds produce different sequences. This means the choice of seed value indirectly affects model performance—some seeds may yield higher accuracy than others due to how they initialize weights or split training data.

**Research Best Practices**

Rather than relying on a single seed, conducting experiments across multiple seeds and averaging results provides statistically robust conclusions. Cross-validation with varied seeds helps isolate genuine algorithm improvements from seed-dependent variance. This repeated sampling approach with controlled randomization enables reproducible research while accounting for the inherent stochasticity of deep learning.​

### T-Test

**T-Test and Its Relevance in Deep Learning**

The t-test is a classical statistical hypothesis test that determines whether two groups have significantly different means. While foundational in statistics, its relevance in deep learning manifests across several distinct domains.

**Core Concept**

The t-test compares the mean of one or two samples against a null hypothesis (typically that means are equal). It produces a t-statistic and p-value, which together indicate whether observed differences are statistically significant or due to random chance. The test assumes data comes from a normal distribution with unknown variance.

**Deep Learning Applications**

**Model Comparison & Evaluation:**

Paired t-tests are extensively used to compare classifier performance. When evaluating two models using k-fold cross-validation, you obtain multiple accuracy measurements. The t-test determines if performance differences are statistically significant—critical for rigorously selecting models beyond simple accuracy comparisons. At significance level p < 0.05, you can confidently claim one model outperforms another.

**Hypothesis Testing & Feature Significance:**

Recent research extends classical statistical testing into deep networks. Gradient-based significance tests assess which input features meaningfully impact neural network predictions, enabling interpretability in black-box models. Additionally, novel frameworks reformulate deep learning problems (like out-of-distribution detection and adversarial robustness) as hypothesis testing problems, combining evidence from entire networks while maintaining type-I error guarantees.

**Two-Sample Testing with Deep Learning:**

Rather than classical t-tests, researchers now learn deep neural network representations optimized for testing whether two datasets come from different distributions—called Deep Maximum Mean Discrepancy (DMMD) and Deep Fisher Discriminant Analysis (DFDA). These data-driven tests adapt to high-dimensional, complex data better than classical parametric tests.

**Floating Point Accuracy:**

In practical deep learning deployment (especially on mobile devices), t-tests and z-tests validate computational accuracy of DNNs, comparing expected vs. actual outputs statistically.

````python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
````

````python
# parameters
n1 = 50   # samples in dataset 1
n2 = 50   # ...and 2
mu1 = 1   # population mean in dataset 1
mu2 = 3   # population mean in dataset 2

# generate the data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)



# plot them
plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2), data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])
plt.show()
````

````python
# t-test via stats package

# _ind = independent samples
t,p = stats.ttest_ind(data1,data2)
print(t)
print(p)
````

````python
# common way to show t-test results in a plot
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) # change the font size

plt.plot(0+np.random.randn(n1)/15, data1, 'ro',markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group 1','Group 2'])

# set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p={p:.3f}')

plt.show()
````

### Derivative

**Derivative and Its Relevance in Deep Learning**

The derivative is a fundamental mathematical concept measuring how a function changes relative to changes in its input. In deep learning, derivatives are the computational backbone enabling neural networks to learn.

**Core Role in Learning**

Derivatives drive **gradient descent**, the optimization algorithm that trains neural networks. During training, derivatives compute the gradient—the vector of partial derivatives of the loss function with respect to each weight and bias. This gradient indicates the direction and magnitude of change needed to minimize error. By following the negative gradient, the network updates parameters to reduce loss iteratively.

**Backpropagation Algorithm**

Backpropagation uses the chain rule of calculus to efficiently compute derivatives throughout the network. It propagates error backward through layers, calculating how each parameter contributes to total loss. The process involves:

1. Forward pass: Computing network output

2. Backward pass: Calculating $\frac{∂C}{∂w}$ and $\frac{∂C}{∂b}$ (partial derivatives of cost with respect to weights and biases) using the chain rule.

3. Parameter updates: Adjusting weights by $w ← w - η*\frac{∂C}{∂w}$, where η is the learning rate.

**Activation Function Derivatives**

The derivative of activation functions (sigmoid, ReLU, etc.) is critical—it directly affects gradient computation in backpropagation. Activation function derivatives multiply through the chain rule across layers, influencing how quickly earlier layers learn. This sensitivity manifests in problems like vanishing gradients in deep networks, where derivatives become exponentially smaller moving backward through layers, preventing weight updates in early layers.

**Practical Implications**

Derivative computation quality determines:

1. Convergence speed: Steeper gradients enable faster learning

2. Learning stability: Poor derivative behavior leads to exploding or vanishing gradients

3. Architecture design: Choosing activation functions with well-behaved derivatives improves training

In essence, derivatives transform the abstract goal of minimizing loss into concrete, computable weight updates, making modern neural network training possible.

````python
import numpy as np
import matplotlib.pyplot as plt

# sympy = symbolic math in Python
import sympy as sym
import sympy.plotting.plot as symplot
````

````python
# repeat with relu and sigmoid

# create symbolic variables in sympy
x = sym.symbols('x')

# create symbolic functions
relu = sym.Max(0,x)
sigmoid = 1 / (1+sym.exp(-x))

# graph the functions
p = symplot(relu,(x,-4,4),label='ReLU',show=False,line_color='blue')
p.extend( symplot(sigmoid,(x,-4,4),label='Sigmoid',show=False,line_color='red') )
p.legend = True
p.title = 'The functions'
p.show()


# graph their derivatives
p = symplot(sym.diff(relu),(x,-4,4),label='df(ReLU)',show=False,line_color='blue')
p.extend( symplot(sym.diff(sigmoid),(x,-4,4),label='df(Sigmoid)',show=False,line_color='red') )
p.legend = True
p.title = 'The derivatives'
p.show()
````
